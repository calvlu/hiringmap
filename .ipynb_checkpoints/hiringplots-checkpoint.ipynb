```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hiringdf = pd.read_csv("aggregate_job_postings_US.csv")
hiringdf.head()
```

```python
hiringdf['date'] = pd.to_datetime(hiringdf['date'])
hiringdf = hiringdf[hiringdf['variable'] == 'total postings']
hiringdf = hiringdf[['date', 'pct_chng_feb_1']]
```

```python
hiringdf
```

```python
#sns.lineplot(x='date', y='pct_chng_feb_1', data=hiringdf)
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(20, 10))
sns.lineplot(x='date', y='pct_chng_feb_1', data=hiringdf)
#plt.xticks(rotation = 90)
plt.title('% change in job postings since Feb 2020, seasonally adjusted, to Mar 2021')
plt.axhline(0, color='0.5', linewidth=2)
#plt.show()
plt.savefig('hiringplot.png')
```

```python
newdf = pd.DataFrame({})
newdf['last'] = hiringdf.groupby([hiringdf['date'].dt.year,hiringdf['date'].dt.month, 'state'],as_index=False)['pct_chng_feb_1'].last()
#hiringdf = hiringdf[hiringdf['date'].dt.is_month_end]
#newdf = newdf.reset_index()
newdf['mean_change'] = hiringdf.groupby([hiringdf['date'].dt.year,hiringdf['date'].dt.month, 'state'])['pct_chng_feb_1'].mean()
#hiringdf.groupby([hiringdf['date'].dt.year,hiringdf['date'].dt.month, 'state'])['pct_chng_feb_1'].last()
newdf#.reset_index(drop=True)
```

```python
sns.lineplot(x='date', y='last', data=newdf)
```
